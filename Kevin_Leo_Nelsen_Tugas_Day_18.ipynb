Soal 1: Visualisasikan distribusi (KDE plot) dari kolom ```bill_length_mm``` dan kolom ```bill_depth_mm``` yang dikelompokkan berdasarkan kolom ```island```. Visualisasi dari ```bill_length_mm``` berada di kiri, sedangkan visualisasi ```bill_depth_mm``` berada di sebelah kanan (sub plots). Buatlah plot se aestetik mungkin. 

Berilah judul pada masing-masing sub-plot, kemudian berilah judul utama di tengah. Judulnya apa? Bebas. Terserah.

````python
df = sns.load_dataset("penguins")
````

````python
df = df.dropna().reset_index(drop = True)
````

````python
df
````

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15,8), 
                                                       sharey = False,
                                                      sharex = False)

# Draw the kde plots
hist_norm = sns.kdeplot(data = df, x ='bill_length_mm', hue = 'island', label = 'Bill Length (mm)', ax = ax1)
hist_logn = sns.kdeplot(data = df, x ='bill_depth_mm',hue='island', label = 'Bill Depth(mm)', ax = ax2)

# Set axis label
ax1.set_ylabel('Density')

ax2.set_ylabel('')

# Set Title for each plots
ax1.set_title('Bill Length')
ax2.set_title('Bill Depth')

fig.suptitle('Density Plots of Bill Dimension', fontsize = 'xx-large')
plt.show()
````

Soal 2: Visualisasikan scatterplot dengan ketentuan sebagai berikut:
- bentuk : subplot dengan 1 column dan 2 row.
- subplot 1:
    - sumbu x : body_mass_g
    - sumbu y : flipper_length_mm
- subplot 2:
    - sumbu x : body_mass_g
    - sumbu y : bill_length_mm
- diwarnai titik2nya berdasarkan island
- buat dengan aestetik
- Kemudian, tulis interpretasi yang Anda peroleh berdasarkan scatterplot tersebut.

````python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10, 10))

sns.scatterplot(x="body_mass_g", y="flipper_length_mm", hue="island" #warna
                , size="body_mass_g" #ukuran titik
                , sizes=(40, 400), alpha=.5, palette="rocket",
            data=df, ax = ax1)

sns.scatterplot(x="body_mass_g", y="bill_length_mm", hue="island" #warna
                , size="body_mass_g" #ukuran titik
                , sizes=(40, 200), alpha=.5, palette="rocket",
            data=df, ax = ax2)

ax1.set_xlabel('')
ax1.set_ylabel('Flipper Length (mm)')

ax2.set_xlabel('Body Mass (grams)')
ax2.set_ylabel('Bill Length (mm)')

fig.suptitle('Body Mass to Body Features plot', fontsize = 'xx-large')
plt.show()
````

Interpretasi: Berdasarkan scatterplot diatas, semakin berat seekor penguin, semakin panjang juga siripnya. penguin jenis Biscoe memiliki rata-rata berat yag lebih dibanding penguin jenis lain.
