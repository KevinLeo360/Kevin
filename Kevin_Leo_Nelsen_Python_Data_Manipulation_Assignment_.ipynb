```python
import pandas as pd
import numpy as np
import re
import string
```

```python
df = pd.read_csv("test.csv")
```

## **Cleaning the DATA**

###***Log Transaction Column***

```python
#I'm defining function to clean the punctuation (') from the data column
def clean(x):
  '''this function is to check the data by looping to every value in the column and 
  remove the punctuation except (.) and split the data to make a list of num type (float) data'''
  for punctuation in string.punctuation.replace(".", ""):
      x = x.replace(punctuation, '')

      split = x.split(" ")
  num_list = [float(num) for num in split]
  return num_list
```

```python
#applying the function to the whole log_transaction columns
df['log_transaction'] = df['log_transaction'].apply(clean)
```

```python
'''I'm using lambda function to make it more effective, I put the round() in the mean and std 
because the decimals is too much and the data turns into "e-6"'''
df['min_trx'] = df['log_transaction'].apply(lambda x: min(x))
df['max_trx'] = df['log_transaction'].apply(lambda x: max(x))
df['avg_trx'] = df['log_transaction'].apply(lambda x: np.mean(x).round(decimals = 3))
df['std_trx'] = df['log_transaction'].apply(lambda x: np.std(x).round(decimals = 3))
df['cnt_trx'] = df['log_transaction'].apply(lambda x: len(x))
```

```python
#I'm checking if the data has to be applied successfully 
df
```

###***Visited City Column***

```python
#fist we should get rid the punctuation
df["visited_city"] = df["visited_city"].str.replace("[^\w\s]",'')
```

```python
#checking the dataframe columns
df["visited_city"]
```

```python
#because the data is already in independent in each value, I'm using str.get_dummies to encode it
visited_city_dummies = df['visited_city'].str.get_dummies(" ")
```

```python
#I add the "visited" prefix to every visited_city column names
visited_city_dummies = visited_city_dummies.add_prefix('visited_')
```

```python
#and finally I merge the result with the current dataframe
df = pd.concat([df, visited_city_dummies], axis=1)
```

### ***Airlines Name Column***

```python
#first, I'm checking the unique value od the names
df['airlines_name'].value_counts()
```

```python
#I'm making a function that returns the unique names to a certain code like "A"
def clean_airlines_name(x):
    if '6c483c0812c96f8ec43bb0ff76eaf716' in x.lower():
        return 'A'
    elif '33199710eb822fbcfd0dc793f4788d30' in x.lower():
        return 'B'
    elif '0a102015e48c1f68e121acc99fca9a05' in x.lower():
        return 'C'
    elif 'ad5bef60d81ea077018f4d50b813153a' in x.lower():
        return 'D'
    elif '74c5549aa99d55280a896ea50068a211' in x.lower():
        return 'E'
    else:
        return 'F'

#and I applying it to the dataframe
df['airlines_name'] = df['airlines_name'].apply(clean_airlines_name)
```
